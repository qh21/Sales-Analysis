# Now let's answer the questions

## How does the number of units sold vary over time?

```python
df['Month'] = df['Date'].dt.to_period('M')
df
```

```python
df['Year'] = df['Date'].dt.to_period('Y')
df
```

```python
df['yearly_total'] = df.groupby('Year')['Units'].transform('sum')
```

```python
df2 = df[['Year', 'yearly_total']].drop_duplicates()
```

```python
df2
```

```python
df2 = df2.reset_index(drop=True)
```

```python
df2
```

```python
sns.lineplot(x=df2['Year'].astype(str), y=df2['yearly_total'])
```

```python
df2 = df2.sort_values(by='Year', ascending=True)
```

```python
df2
```

```python
sns.lineplot(x=df2['Year'].astype(str), y=df2['yearly_total'])
plt.xlabel('Year')
plt.ylabel('Total Sold Units')
plt.title('Total Sold Units per Year')
```

```python
df3 = df.groupby('Zip')['Units'].sum().reset_index()
df3 = df3.sort_values(by='Units', ascending=False)

# Display the top zip codes
print("Top Zip Codes by Total Units Sold:\n\n")
print(df3.head(10))
```

## Is there a noticeable trend in revenue over the given dates?

```python
df4 = df.groupby('Date')['Revenue'].sum().reset_index()

# Time Series Plot with Aggregated Data
plt.figure(figsize=(20, 12))
plt.plot(df4['Date'], df4['Revenue'], marker='o', linestyle='-', color='b')
plt.title('Aggregated Revenue Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

## Are there any specific dates or months where sales are consistently higher or lower?

```python
# Aggregate data by month and year
monthly_aggregated = df.groupby(['Year', 'Month']).agg({'Revenue': 'sum', 'Units': 'sum'}).reset_index()

# Visualize the monthly revenue trend
plt.figure(figsize=(30, 12))
plt.plot(monthly_aggregated['Year'].astype(str) + '-' + monthly_aggregated['Month'].astype(str), monthly_aggregated['Revenue'], marker='o', linestyle='-', color='b')
plt.title('Monthly Revenue Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90, ha='right')
plt.grid(True)
plt.show()
```

## What are the sales patterns across different zip codes?

```python
# Aggregate data by zip code
zip_code_aggregated = df.groupby('Zip').agg({'Revenue': 'sum', 'Units': 'sum'}).reset_index()

# Sort the data by total revenue to highlight top-contributing zip codes
zip_code_aggregated = zip_code_aggregated.sort_values(by='Revenue', ascending=False)

# Select the top N zip codes (adjust N based on your needs)
top_n = 10
top_zip_codes = zip_code_aggregated.head(top_n)

# Visualize the sales patterns for the top N zip codes
plt.figure(figsize=(12, 6))
plt.bar(top_zip_codes['Zip'].astype(str), top_zip_codes['Revenue'], color='b')
plt.title(f'Top {top_n} Sales Patterns Across Zip Codes')
plt.xlabel('Zip Code')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

## Is there a correlation between the zip code and the number of units sold or revenue generated?

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Zip', 'Units', 'Revenue']].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix: Zip Code, Units Sold, Revenue')
plt.show()
```

## Are there specific countries that contribute significantly to the sales?

```python
# Aggregate data by country
country_aggregated = df.groupby('Country').agg({'Revenue': 'sum', 'Units': 'sum'}).reset_index()

# Sort the data by total revenue to highlight top-contributing countries
country_aggregated = country_aggregated.sort_values(by='Revenue', ascending=False)

# Visualize the sales contribution by country
plt.figure(figsize=(12, 6))
plt.bar(country_aggregated['Country'], country_aggregated['Revenue'], color='b')
plt.title('Sales Contribution by Country')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

## How does the performance of ProductID 1082 compare to other products in the dataset?

```python
# Filter data for ProductID 1082
product_1082 = df[df['ProductID'] == 1082]

# Aggregate data for ProductID 1082
product_1082_aggregated = product_1082.groupby('ProductID').agg({'Units': 'sum', 'Revenue': 'sum'}).reset_index()

# Aggregate data for all products (excluding ProductID 1082)
other_products_aggregated = df[df['ProductID'] != 1082].groupby('ProductID').agg({'Units': 'sum', 'Revenue': 'sum'}).reset_index()

# Compare total units sold
plt.bar(['Product 1082', 'Other Products'], [product_1082_aggregated['Units'].values[0], other_products_aggregated['Units'].sum()])
plt.title('Comparison of Total Units Sold')
plt.ylabel('Total Units Sold')
plt.show()

# Compare total revenue generated
plt.bar(['Product 1082', 'Other Products'], [product_1082_aggregated['Revenue'].values[0], other_products_aggregated['Revenue'].sum()])
plt.title('Comparison of Total Revenue Generated')
plt.ylabel('Total Revenue')
plt.show()

# Compare average revenue per unit
average_revenue_per_unit_1082 = product_1082_aggregated['Revenue'].values[0] / product_1082_aggregated['Units'].values[0]
average_revenue_per_unit_other = other_products_aggregated['Revenue'].sum() / other_products_aggregated['Units'].sum()

plt.bar(['Product 1082', 'Other Products'], [average_revenue_per_unit_1082, average_revenue_per_unit_other])
plt.title('Comparison of Average Revenue per Unit')
plt.ylabel('Average Revenue per Unit')
plt.show()
```

## Is there a consistent pricing strategy for ProductID 1082, given the constant revenue for each unit sold?

```python
# Filter data for ProductID 1082
product_1082 = df[df['ProductID'] == 1082]

# Check if there is a constant pricing strategy
constant_revenue_per_unit = product_1082['Revenue'].nunique() == 1

# Visualize the relationship between Units and Revenue
plt.scatter(product_1082['Units'], product_1082['Revenue'])
plt.title('Relationship between Units and Revenue for ProductID 1082')
plt.xlabel('Units')
plt.ylabel('Revenue')
plt.show()

# Print information about pricing strategy
if constant_revenue_per_unit:
    print("There is a consistent pricing strategy for ProductID 1082.")
else:
    print("The pricing strategy for ProductID 1082 is not consistent.")
```

## Do certain countries prefer this product  (ProductID 1082) more than others?

```python
# Filter data for ProductID 1082
product_1082 = df[df['ProductID'] == 1082]

# Aggregate data for ProductID 1082 by country
product_1082_by_country = product_1082.groupby('Country').agg({'Units': 'sum', 'Revenue': 'sum'}).reset_index()

# Sort the data by total units sold to highlight top-contributing countries
product_1082_by_country = product_1082_by_country.sort_values(by='Units', ascending=False)

# Visualize sales distribution by country
plt.figure(figsize=(12, 6))
plt.bar(product_1082_by_country['Country'], product_1082_by_country['Units'], color='b')
plt.title('Sales Distribution of ProductID 1082 by Country')
plt.xlabel('Country')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

##  Are there variations in sales performance between different countries?


```python
country_aggregated = df.groupby('Country').agg({'Units': 'sum', 'Revenue': 'sum'}).reset_index()

# Sort the data by total revenue to highlight top-contributing countries
country_aggregated = country_aggregated.sort_values(by='Revenue', ascending=False)

# Visualize sales performance by country
plt.figure(figsize=(12, 6))
plt.bar(country_aggregated['Country'], country_aggregated['Revenue'], color='b')
plt.title('Sales Performance by Country')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

## What is the overall performance of ProductID 1082 in terms of units sold and revenue generated?

```python
# Filter data for ProductID 1082
product_1082 = df[df['ProductID'] == 1082]

# Calculate overall performance metrics
total_units_sold = product_1082['Units'].sum()
total_revenue_generated = product_1082['Revenue'].sum()
average_revenue_per_unit = total_revenue_generated / total_units_sold if total_units_sold > 0 else 0

# Print overall performance metrics
print(f"Total Units Sold for ProductID 1082: {total_units_sold}")
print(f"Total Revenue Generated for ProductID 1082: {total_revenue_generated}")
print(f"Average Revenue per Unit for ProductID 1082: {average_revenue_per_unit}")
plt.figure(figsize=(12, 6))
# Visualize overall performance
plt.bar(['Total Units Sold', 'Total Revenue Generated', 'Average Revenue per Unit'],
        [total_units_sold, total_revenue_generated, average_revenue_per_unit])
plt.title('Overall Performance of ProductID 1082')
plt.ylabel('Value')
plt.show()
```
